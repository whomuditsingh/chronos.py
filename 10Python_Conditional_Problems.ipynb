## 1. E-Commerce Discount Calculator
### Apply discounts based on membership tier and total cart value.

```python
customer_data = {
    "name": "Alice",
    "cart_total": 150.00,
    "membership": "Premium" # Options: None, Basic, Premium
}

total = customer_data["cart_total"]
if customer_data["membership"] == "Premium":
    final_price = total * 0.80 if total > 100 else total * 0.90
elif customer_data["membership"] == "Basic":
    final_price = total * 0.95
else:
    final_price = total

print(f"Final price for {customer_data['name']}: ${final_price:.2f}")
```

## 2. Flight Baggage Fee Checker
### Determine baggage fees based on weight and ticket class.

```python
passenger = {
    "ticket_class": "Economy", # Options: Economy, Business, First
    "bag_weight_kg": 28
}

weight = passenger["bag_weight_kg"]
t_class = passenger["ticket_class"]
fee = 0

if t_class == "Economy":
    if weight > 23:
        fee = 50 + ((weight - 23) * 10) # Base fee + $10 per extra kg
elif t_class == "Business":
    if weight > 32:
        fee = (weight - 32) * 15
# First class flies free regardless of weight up to a limit (ignored here)

print(f"Baggage fee: ${fee}")
```

## 3. Library Fine System
### Calculate fines based on days overdue and the type of book.

```python
book_return = {
    "title": "Python Crash Course",
    "days_overdue": 5,
    "type": "Reference" # Options: Regular, Reference, Media
}

days = book_return["days_overdue"]
b_type = book_return["type"]
fine = 0.0

if days > 0:
    if b_type == "Reference":
        fine = days * 2.00 # High penalty for reference books
    elif b_type == "Media":
        fine = days * 1.50
    else:
        fine = days * 0.50

print(f"Overdue fine: ${fine:.2f}")
```

## 4. Bank Loan Approval System
### Approve or reject a loan based on credit score and debt-to-income (DTI) ratio.

```python
applicant = {
    "credit_score": 720,
    "dti_ratio": 0.35, # 35% debt-to-income
    "income": 80000
}

score = applicant["credit_score"]
dti = applicant["dti_ratio"]

if score >= 750 and dti <= 0.40:
    status = "Automatically Approved"
elif 650 <= score < 750 and dti <= 0.35:
    status = "Approved with Standard Rates"
elif 600 <= score < 650:
    status = "Manual Underwriting Required"
else:
    status = "Declined"

print(f"Loan Status: {status}")
```

## 5. Health Insurance Premium Estimator
### Calculate monthly premiums based on age, smoking status, and pre-existing conditions.

```python
user_profile = {
    "age": 45,
    "is_smoker": True,
    "pre_existing_conditions": 1
}

base_premium = 200

if user_profile["age"] > 50:
    base_premium += 100
elif user_profile["age"] > 30:
    base_premium += 50

if user_profile["is_smoker"]:
    base_premium *= 1.5 # 50% surcharge

if user_profile["pre_existing_conditions"] > 0:
    base_premium += (user_profile["pre_existing_conditions"] * 30)

print(f"Estimated Monthly Premium: ${base_premium:.2f}")
```

## 6. Smart Home HVAC Controller
### Determine whether the AC or Heater should turn on based on the current temp, target temp, and mode.

```python
thermostat = {
    "current_temp": 78,
    "target_temp": 72,
    "system_mode": "Auto" # Options: Auto, Cool, Heat, Off
}

curr = thermostat["current_temp"]
target = thermostat["target_temp"]
mode = thermostat["system_mode"]

if mode == "Off":
    action = "System Idle"
elif mode == "Cool" or (mode == "Auto" and curr > target + 2):
    action = "Cooling On"
elif mode == "Heat" or (mode == "Auto" and curr < target - 2):
    action = "Heating On"
else:
    action = "Fan Only"

print(f"HVAC Action: {action}")
```

## 7. Restaurant Tip Recommender
### Suggest a tip percentage based on service rating and party size (automatic gratuity for large parties).

```python
dining_experience = {
    "bill_total": 120.50,
    "service_rating": 4, # Scale 1-5
    "party_size": 7
}

if dining_experience["party_size"] >= 6:
    tip_percent = 0.18 # Auto gratuity
    print("Automatic 18% gratuity applied for large party.")
else:
    rating = dining_experience["service_rating"]
    if rating == 5:
        tip_percent = 0.20
    elif rating == 4:
        tip_percent = 0.15
    elif rating >= 2:
        tip_percent = 0.10
    else:
        tip_percent = 0.05

tip_amount = dining_experience["bill_total"] * tip_percent
print(f"Suggested Tip: ${tip_amount:.2f}")
```

## 8. Software Subscription Access
### Check if a user can access a specific feature based on their role and license status.

```python
user_session = {
    "username": "dev_guru",
    "role": "Editor", # Options: Admin, Editor, Viewer
    "license_status": "Active",
    "feature_requested": "Delete_Database"
}

if user_session["license_status"] != "Active":
    access = "Denied: License Expired"
else:
    role = user_session["role"]
    req = user_session["feature_requested"]
    
    if role == "Admin":
        access = "Granted"
    elif role == "Editor" and req != "Delete_Database":
        access = "Granted"
    elif role == "Viewer" and req == "Read_Only":
        access = "Granted"
    else:
        access = "Denied: Insufficient Permissions"

print(f"Access Status: {access}")
```

## 9. Hotel Room Upgrader
### Determine if a guest gets a free room upgrade based on loyalty points and current availability.

```python
hotel_system = {
    "guest_loyalty_points": 15000,
    "booked_room": "Standard",
    "suites_available": 3
}

points = hotel_system["guest_loyalty_points"]
available = hotel_system["suites_available"]

if available > 0:
    if points >= 20000:
        upgrade = "Presidential Suite"
    elif points >= 10000:
        upgrade = "Junior Suite"
    else:
        upgrade = "No Upgrade"
else:
    upgrade = "No Upgrade (Fully Booked)"

print(f"Room Allocation: {upgrade}")
```

## 10. Tax Bracket Estimator (Simplified)
### Calculate estimated tax owed based on gross income and filing status.

```python
taxpayer = {
    "gross_income": 95000,
    "filing_status": "Single" # Options: Single, Married
}

income = taxpayer["gross_income"]
status = taxpayer["filing_status"]
tax_owed = 0

if status == "Single":
    if income <= 40000:
        tax_owed = income * 0.12
    elif income <= 85000:
        tax_owed = 4800 + ((income - 40000) * 0.22)
    else:
        tax_owed = 14700 + ((income - 85000) * 0.24)
elif status == "Married":
    # Wider brackets for married couples
    if income <= 80000:
        tax_owed = income * 0.12
    else:
        tax_owed = 9600 + ((income - 80000) * 0.22)

print(f"Estimated Tax Owed: ${tax_owed:,.2f}")
```
