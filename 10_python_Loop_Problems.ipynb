1. Grouping Sales Data by Category
Iterate through a list of transaction dictionaries to calculate total revenue per product category.

```python
transactions = [
    {"id": 1, "category": "Electronics", "amount": 250.0},
    {"id": 2, "category": "Clothing", "amount": 45.5},
    {"id": 3, "category": "Electronics", "amount": 120.0},
    {"id": 4, "category": "Home", "amount": 75.0},
    {"id": 5, "category": "Clothing", "amount": 60.0}
]

revenue_by_category = {}
for txn in transactions:
    cat = txn["category"]
    # Initialize or add to the existing category
    revenue_by_category[cat] = revenue_by_category.get(cat, 0) + txn["amount"]

print("Revenue by Category:", revenue_by_category)
```

2. Calculating a Moving Average
Use a loop to calculate a 3-day moving average over a series of stock prices

```python
stock_prices = [120, 122, 121, 125, 128, 126, 130, 131]
window_size = 3
moving_averages = []

# Loop up to the point where a full window can still be formed
for i in range(len(stock_prices) - window_size + 1):
    window = stock_prices[i : i + window_size]
    average = sum(window) / window_size
    moving_averages.append(round(average, 2))

print("3-Day Moving Averages:", moving_averages)
```

3. Flattening a 2D Matrix
Convert a 2D grid (list of lists) into a single, flat 1D list using nested loops.

```python
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flat_list = []
for row in grid:
    for item in row:
        flat_list.append(item)

print("Flattened List:", flat_list)
# Note: Can also be done with list comprehension: [item for row in grid for item in row]
```

4. Finding the Top Performing Student
Iterate through a dictionary mapping students to their grades to find the student with the highest average.

```python
student_grades = {
    "Alice": [85, 90, 92],
    "Bob": [78, 80, 85],
    "Charlie": [95, 92, 98],
    "Diana": [88, 89, 90]
}

top_student = ""
highest_avg = 0.0

for student, grades in student_grades.items():
    avg = sum(grades) / len(grades)
    if avg > highest_avg:
        highest_avg = avg
        top_student = student

print(f"Top Student: {top_student} (Average: {highest_avg:.2f})")
```

5. Inventory Discrepancy Checker
Use zip() in a loop to compare an expected inventory list against a counted inventory list.

```python
items = ["Laptops", "Monitors", "Keyboards", "Mice"]
expected_counts = [50, 120, 300, 250]
actual_counts = [48, 120, 295, 250]

discrepancies = []

for item, expected, actual in zip(items, expected_counts, actual_counts):
    if expected != actual:
        discrepancies.append({
            "item": item,
            "missing": expected - actual
        })

print("Inventory Discrepancies:", discrepancies)
```

6. Building an Inverted Index
Simulate how search engines work by mapping words back to the document IDs they appear in.

```python
documents = {
    "doc_1": "python is great for data processing",
    "doc_2": "data processing requires loops",
    "doc_3": "loops in python are powerful"
}

inverted_index = {}

for doc_id, text in documents.items():
    words = text.split()
    for word in words:
        if word not in inverted_index:
            inverted_index[word] = set()
        inverted_index[word].add(doc_id)

print("Inverted Index for 'data':", inverted_index.get("data"))
```

7. Data Paginator (Chunking)
Use a while or for loop with a step to slice a large dataset into smaller "pages" or chunks.

```python
user_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
page_size = 3
pages = []

for i in range(0, len(user_ids), page_size):
    chunk = user_ids[i : i + page_size]
    pages.append(chunk)

for page_num, page in enumerate(pages, 1):
    print(f"Page {page_num}: {page}")
```

8. Simulated File Download (While Loop with Progress)
Simulate downloading a file in chunks using a while loop until a threshold is met.

```python
import random

total_file_size_mb = 50
downloaded_mb = 0
attempts = 0

print("Starting download...")
while downloaded_mb < total_file_size_mb:
    # Simulate network speed varying between 2 and 8 MB/s
    chunk = random.randint(2, 8) 
    downloaded_mb += chunk
    attempts += 1
    
    # Cap the downloaded amount at the total size
    if downloaded_mb > total_file_size_mb:
        downloaded_mb = total_file_size_mb
        
    print(f"Attempt {attempts}: Downloaded {downloaded_mb}/{total_file_size_mb} MB")

print("Download complete!")
```

9. Filtering Deeply Nested Logs
Extract specific error messages from a list of server objects, each containing its own log lists.

```python
servers = [
    {"name": "Web-01", "logs": [{"level": "INFO", "msg": "Started"}, {"level": "ERROR", "msg": "Timeout"}]},
    {"name": "DB-01", "logs": [{"level": "INFO", "msg": "Query OK"}, {"level": "WARN", "msg": "High Load"}]},
    {"name": "API-01", "logs": [{"level": "ERROR", "msg": "Auth Failed"}, {"level": "ERROR", "msg": "500 Internal"}]}
]

error_summary = []

for server in servers:
    for log in server["logs"]:
        if log["level"] == "ERROR":
            error_summary.append(f"{server['name']} - {log['msg']}")

print("Critical Errors Found:")
for err in error_summary:
    print("-", err)
```

10. Matrix Transposition
Swap the rows and columns of a matrix using nested loops.

```python
matrix = [
    [1, 2],
    [3, 4],
    [5, 6]
]

# The original is 3x2, the transposed should be 2x3
rows, cols = len(matrix), len(matrix[0])
transposed = [[0 for _ in range(rows)] for _ in range(cols)]

for i in range(rows):
    for j in range(cols):
        transposed[j][i] = matrix[i][j]

print("Original Matrix:")
for row in matrix: print(row)

print("\nTransposed Matrix:")
for row in transposed: print(row)
```
